11. Line Chart: Plotting the Closing Prices over Time

```python
fig = px.line(goldman_sachs_df, x='Date', y='Close', title='Closing Prices over Time')
fig.show()
```

12. Area Chart: Volume Traded over Time

```python
fig1 = px.area(goldman_sachs_df, x='Date', y='Volume', title='Volume Traded over Time')
fig1.show()
```

13. Bar Chart: Dividends Paid per Date

```python
fig3 = px.bar(goldman_sachs_df, x='Date', y='Dividends', title='Dividends Paid per Date')
fig3.show()
```

14. Box Plot: Distribution of Closing Prices by Month

```python
goldman_sachs_df['Month'] = (goldman_sachs_df['Date']).dt.month
goldman_sachs_df['Year'] = (goldman_sachs_df['Date']).dt.year

fig4 = px.box(goldman_sachs_df, x='Month', y='Close', title='Distribution of Closing Prices by Month')
fig4.show()
```

15. Scatter Plot : Opening vs. Closing Prices

```python
fig5 = px.scatter(goldman_sachs_df, x='Open', y='Close', title='Opening vs. Closing Prices')
fig5.show()
```
